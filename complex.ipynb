```python
def sinplot(flip=2):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
def quadratic(params):
    a, b, c = params
    def f(x):
        return a * x**2 + b * x + c
    return f
```

```python
params = 1.0, -4.0, 1.0
```

```python
p = quadratic(params)
```

```python
x = np.linspace(-10.0, 10.0, 1000)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(x, p(x))
```

```python
def poly_to_params(p):
    c = p(0.0)
    apb = p(1.0) - c
    amb = p(-1.0) - c
    a = (apb + amb) / 2.0
    b = apb - a
    return (a, b, c)
    
```

```python
def quadratic_solve(params):
    a, b, c = params
    s = (b**2 - 4.0 * a * c)**0.5
    x = [(-b + s) / (2.0 * a), (-b - s) / (2.0 * a)]
    return x
```

```python
x1, x2 = quadratic_solve(params)
x1, x2
```
